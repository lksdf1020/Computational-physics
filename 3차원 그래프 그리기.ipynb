# 3차원 그래프 그리기

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(6, 6)) # figsize로 사이즈 조절
ax = fig.add_subplot(111, projection='3d') 
x = [1,2,3,4,5] # x,y,z해당 데이터셋, 행렬 형태 상관없음
y = [6,3,1,2,7]
z = [7,1,2,3,8]
color = [1,1,1,0,0] # 색상 설정
ax.scatter(x,y,z,c=color)
```

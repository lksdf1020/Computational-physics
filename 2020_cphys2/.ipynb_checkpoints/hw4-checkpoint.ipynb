```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
Iris = datasets.load_iris()
```

```python
Iris.target
```

```python
Iris.feature_names
```

```python
X = Iris.data
y = Iris.target
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

##### 그래프 1. 꽃받침 길이(X-axis), 꽃받침 너비(Y-axis) 에 따른 꽃의 분류
##### setosa(보라색), versicolor(초록색) virginica(노란색)


- sepal의 크기는 setosa 그룹에서는 뚜렷하게 구분되나, 나머지 두 그룹에서는 뚜렷히 구분되지 않는다.

```python
plt.scatter(X[:,2],X[:,3],c=y)
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.show()
```

#### 그래프 2. 꽃잎의 길이(X-axis), 꽃잎의 너비(Y-axis) 에 따른 꽃의 분류

- sepal과 달리 명확하게 상대적으로 세 그룹이 petal 분포에선 구분된다.

# Logistic Regression 

- Setosa 그룹의 데이터가 Sepal과 Petal 두 그룹에서 명확하게 나뉘므로, Setosa그룹과 아닌 그룹으로 구분하였다. 

```python
from sklearn.linear_model import LogisticRegression
```

```python
X1_sepal = [[i] for i in X[:,0]] # X[:,i], i=0,1,2,3(sepal length, sepal width, patal length, petal width)
y1 = [i if i == 0 and 1 else 1 for i in y]
```

```python
log_reg_S = LogisticRegression(solver = 'lbfgs')
log_reg_S.fit(X1_sepal,y1)
```

```python
log_reg_S.coef_, log_reg_S.intercept_
```

```python
x = [[i] for i in np.linspace(4,8,100)]
plt.plot(X1_sepal,y1,'o')
plt.plot(x,log_reg_S.predict(x))
```

##### 그래프 3. 꽃받침의 길이에 따른 Logistic Regression function(주황색 실선)

- Sapal 분포의 경우 그래프 1에서 예측할 수 있듯이, 그래프가 몇몇 데이터를 포함하지 못한다.

- Petal 값에 따른 분포는 그래프 2에서 볼 수 있듯이, 뚜렷하게 두 그룹이 나뉘므로, 모든 데이터가 함수에 포함 될 것이다. 또한 Score값도 더 높을 것으로 예상된다.

```python
X1_petal = [[i] for i in X[:,3]] # X[:,i], i=0,1,2,3(sepal length, sepal width, patal length, petal width)
```

```python
log_reg_P = LogisticRegression(solver = 'lbfgs')
log_reg_P.fit(X1_petal,y1)
```

```python
log_reg_P.coef_, log_reg_P.intercept_
```

```python
x = [[i] for i in np.linspace(-0,3,100)]
plt.plot(X1_petal,y1,'o')
plt.plot(x,log_reg_P.predict(x))
```

##### 그래프 4. 꽃잎의 길이에 따른 Logistic Regression function(주황색 실선)

```python
log_reg_S.score(X1_sepal,y1)
```

```python
log_reg_P.score(X1_petal,y1)
```

### 보유한 데이터에 의하면 Sepal 보다 Petal 크기가 Setosa를 분류하는 데 더 좋은 factor이다.

# Support Vector Machine

```python
from sklearn.svm import SVC
```

```python
svm = SVC(kernel = 'rbf',gamma = 'auto')
svm.fit(X,y)
plt.scatter(X[:,0],X[:,1],c=svm.predict(X))
```

##### 그래프 5. SVM을 기반으로 한 꽃받침 길이(X-axis), 꽃받침 너비(Y-axis) 에 따른 꽃의 분류
##### setosa(보라색), versicolor(초록색) virginica(노란색)


```python
svm = SVC(kernel = 'rbf',gamma = 'auto')
svm.fit(X,y)
plt.scatter(X[:,2],X[:,3],c=svm.predict(X))
```

##### 그래프 6. SVM을 기반으로 한 꽃잎 길이(X-axis), 꽃잎 너비(Y-axis) 에 따른 꽃의 분류
##### setosa(보라색), versicolor(초록색) virginica(노란색)


```python
svm.score(X,y)
```

#### 실제 분류와 SVC를 통한 예측된 꽃의 종류는 거의 동일함을 확인할 수 있다.
